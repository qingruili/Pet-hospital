```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from statsmodels.stats.multicomp import pairwise_tukeyhsd
from scipy import stats
```

```python
data = pd.read_csv("Clean_Data/hotel_clean.csv")
```

```python
num_1 = data[data['num_bedroom'] == 1]['price']
num_2 = data[data['num_bedroom'] == 2]['price']
num_3 = data[data['num_bedroom'] == 3]['price']
num_4 = data[data['num_bedroom'] == 4]['price']
num_5_or_more = data[data['num_bedroom'] >= 5]['price']
```

```python
anova = stats.f_oneway(num_1, num_2, num_3, num_4,num_5_or_more)
print(anova)
print(anova.pvalue)
```

```python
data2 = data.copy()
data2['num_bedroom'] = np.where(data2['num_bedroom'] >=5, '5 or more',data2['num_bedroom'])
```

```python
posthoc = pairwise_tukeyhsd(
    data2['price'], data2['num_bedroom'],
    alpha=0.05)
```

```python
print(posthoc)
```

```python
fig = posthoc.plot_simultaneous()
```
