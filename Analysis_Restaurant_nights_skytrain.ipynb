```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
df = pd.read_csv('Clean_Data/hotel_clean.csv')
restaurant = df.loc[:, ['id', 'price', 'num_restaurant']]
restaurant
```

```python
print(restaurant.describe())
```

```python
fit = linregress(restaurant['num_restaurant'], restaurant['price'])
fit.pvalue
restaurant['pred_price'] = restaurant['num_restaurant'] * fit.slope + fit.intercept
```

```python

plt.figure(figsize=(10, 6))

plt.xticks(rotation=25)
plt.plot(restaurant['num_restaurant'], restaurant['price'], 'b.', alpha=0.5)
plt.plot(restaurant['num_restaurant'], restaurant['pred_price'], 'r-', linewidth=3)
plt.xlabel('Number of Restaurants')
plt.ylabel('Price')
plt.title('Linear Regression: Number of Restaurants vs. Price')
plt.show()
```

```python
correlation = restaurant[['num_restaurant', 'price']].corr()
print(correlation)
```
